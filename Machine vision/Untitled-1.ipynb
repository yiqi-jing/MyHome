```python
# cv2 是 OpenCV 库，用于图像和视频处理。
# matplotlib.pyplot 是用于绘图和可视化的库，常用于显示图像和绘制图表。
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img_path = r"F:\Python\Data\quality_restoration_20250709003543265.jpg"
img = cv2.imread(img_path, 1)
plt.imshow(img)
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2Lab))
```

```python

hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
plt.imshow(hsv_img)
```

```python
alpha = 1.5  # 对比度控制 (1.0-3.0)
beta = 0    # 亮度控制 (0-100)
adjusted = cv2.convertScaleAbs(img, alpha=alpha, beta=beta)
plt.imshow(adjusted)
```

```python
brightened = cv2.convertScaleAbs(img, alpha=1, beta=50)
plt.imshow(brightened)
```

```python
hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
hsv_img[:, :, 1] = cv2.multiply(hsv_img[:, :, 1], 1.5)  # 增加饱和度
saturated = cv2.cvtColor(hsv_img, cv2.COLOR_HSV2BGR)
plt.imshow(saturated)
```

```python
gamma = 1.2  # 伽马值
gamma_corrected = cv2.LUT(img, np.array([((i / 255.0) ** gamma) * 255 for i in np.arange(0, 256)]).astype("uint8"))
plt.imshow(gamma_corrected)
```
