```python
# 使用xpath解析数据
# 把html文本转成一棵树
htree = etree.HTML(res.text)
# 电影名称：
# /html/body/div[2]/div/div[2]/div/ul/li[1]/h3/a/span[1]
# /html/body/div[2]/div/div[2]/div/ul/li[2]/h3/a/span[1]


# 电影介绍：
# /html/body/div[2]/div/div[2]/div/ul/li[1]/h3/a/span[2]
# /html/body/div[2]/div/div[2]/div/ul/li[2]/h3/a/span[2]


# 上映时间：
# /html/body/div[2]/div/div[2]/div/ul/li[1]/h3/a/span[3]
# /html/body/div[2]/div/div[2]/div/ul/li[2]/h3/a/span[3]


# 主演：
# /html/body/div[2]/div/div[2]/div/ul/li[1]/h3/a/span[5]
# /html/body/div[2]/div/div[2]/div/ul/li[2]/h3/a/span[5]


# 背景:
# /html/body/div[2]/div/div[2]/div/ul/li[1]/h3/a/span[6]
# /html/body/div[2]/div/div[2]/div/ul/li[2]/h3/a/span[6]


# 封面:<img src="//image11.m1905.cn/uploadfile/2017/0220/thumb_1_265_376_20170220093804152476.jpg" class="lazyImg" width="265" height="376" alt="魔鬼部队" style="">
# /html/body/div[2]/div/div[2]/div/ul/li[1]/a/img
# /html/body/div[2]/div/div[2]/div/ul/li[2]/a/img


# 评分:
# /html/body/div[2]/div/div[2]/div/ul/li[1]/a/span[2]
# /html/body/div[2]/div/div[2]/div/ul/li[2]/a/span[2]


movie_items = htree.xpath('/html/body/div[2]/div/div[2]/div/ul/li')

# 循环标题，记录电影从第一个开始
for index, item in enumerate(movie_items, start=1):


    # 电影名称
    name_list = item.xpath('.//h3/a/span[1]//text()')
    movie_name = name_list[0].strip() if name_list else "暂无名称"
    
    # 电影介绍
    intro_list = item.xpath('.//h3/a/span[2]//text()')
    movie_intro = intro_list[0].strip() if intro_list else "暂无介绍"
    
    # 上映时间
    time_list = item.xpath('.//h3/a/span[3]//text()')
    release_time = time_list[0].strip() if time_list else "未公布"
    
    # 主演
    actors_list = item.xpath('.//h3/a/span[5]//text()')
    actors = actors_list[0].strip() if actors_list else "暂无信息"
    
    # 背景
    bg_list = item.xpath('.//h3/a/span[6]//text()')
    background = bg_list[0].strip() if bg_list else "暂无背景介绍"
    
    # 评分
    rating_list = item.xpath('.//a/span[@class="score"]/text()')
    movie_rating = rating_list[0].strip() if rating_list else "暂无评分"
    
    # 封面
    cover_list = item.xpath('.//a/img/@src')
    cover_url = cover_list[0] if cover_list else "暂无封面"

    title = f"电影{index}:"


    print(title)
    print(f"名称: <<{movie_name}>>")
    print(f"介绍: {movie_intro}")
    print(f"上映: {release_time}")
    print(f"主演: {actors}")
    print(f"背景: {background}")
    print(f"评分: {movie_rating}")
    print(f"封面: {cover_url}\n")
```
