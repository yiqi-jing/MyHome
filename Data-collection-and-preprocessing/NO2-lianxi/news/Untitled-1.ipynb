```python
# 导入必要的库
import requests as req
import re
import json
```

```python
#提取News2.txt中的内容,为列表格式
with open('F://Source-code-management-repository/MyHome/Data-collection-and-preprocessing/Data/news2.txt', 'r', encoding='GBK', errors='replace') as f:
    
    Data = ''
    
    line = f.readline()
    while len(line) != 0:
        Data = Data + line
        line = f.readline()
    print(Data)
```

```python
# 将上述列表格式转换为json格式
news = json.loads(Data)
print(news['hotsearch'])
```

```python
# 将json文件中字典中的"card_title"一一提取出来，并打印储存
data = news['hotsearch']
for i in range(len(data)):
    print(data[i]['card_title'])
```

```python

for new in news['hotsearch']:
    print(new['card_title'])
```

```python
# 将数据写入文件
with open('news.txt', 'a+', encoding='utf-8') as f:
    for news in data:
        f.write(news['card_title'] + '\n')
```
